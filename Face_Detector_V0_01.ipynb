# **FACE DETECTOR WITH OPEN CV** 



### **MENGIMPORT LIBLARY OPENCV** 

```python
import cv2 #cv2 adalah singkatan dari OpenCV (Open Source Computer Vision Library). 
#Library ini menyediakan berbagai fungsi dan algoritma untuk memanipulasi gambar dan video, 
#termasuk pendeteksian objek seperti wajah
import numpy as np #numpy adalah library yang sering digunakan dalam pemrosesan data numerik. 
#Di sini, kita mengimpor numpy untuk digunakan dalam beberapa operasi pada gambar.
```

### **MENGUPLOAD FOTO** 

```python
from google.colab import files 
#mengimpor modul files dari paket google.colab. 
#Modul ini menyediakan fungsi untuk mengunggah dan mengunduh file dalam lingkungan Google Colab.

uploaded = files.upload() #digunakan untuk mengunggah file dari komputer Anda ke Google Colab. 
#Setelah mengeksekusi baris ini, Anda akan melihat tampilan untuk memilih file dari komputer Anda dan mengunggahnya ke sesi Google Colab.
```

### **MEMBACA GAMBAR** 

```python
image_path = next(iter(uploaded))
#digunakan untuk mendapatkan path (lokasi) gambar yang telah diunggah.
#uploaded artinya dictionary yang berisi file yang diunggah, 
#dan next(iter()) digunakan untuk mendapatkan path dari file yang pertama kali diunggah.

image = cv2.imread(image_path)
#digunakan untuk membaca gambar dari path yang diberikan. 
#Fungsi imread() mengembalikan gambar dalam bentuk matriks piksel.
```



*   Kode di atas bertujuan untuk membaca gambar yang telah diunggah menggunakan OpenCV.
*   Setelah kode ini dieksekusi, gambar yang diunggah akan dibaca dan disimpan dalam variabel image sebagai matriks piksel.

### **MEMBUAT DETECTOR WAJAH** 

```python
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
#digunakan untuk membuat objek detektor wajah. 
#Kode cv2.data.haarcascades + 'haarcascade_frontalface_default.xml'merujuk pada path file XML yang berisi informasi 
#tentang deteksi wajah menggunakan metode Haar cascades. 
#File ini biasanya disertakan dengan instalasi OpenCV.

faces = face_cascade.detectMultiScale(image, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))
#face_cascade.detectMultiScale(image, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30)) digunakan untuk mendeteksi wajah dalam gambar. 
#Metode detectMultiScale memerlukan parameter gambar yang akan dianalisis. 
#Parameter lainnya adalah scaleFactor, minNeighbors, dan minSize yang mengatur parameter untuk proses deteksi wajah. 
#Hasil deteksi wajah akan disimpan dalam variabel faces.
```



*   Kode di atas bertujuan untuk membuat objek detektor wajah menggunakan metode Haar cascades yang disediakan oleh OpenCV.

### **MENGGAMBAR KOTAK PADA WAJAH** 

```python
for (x, y, w, h) in faces:
    cv2.rectangle(image, (x, y), (x+w, y+h), (0, 255, 0), 2)
    #Melalui perulangan for, kita iterasi melalui setiap wajah yang dideteksi dalam faces
    #cv2.rectangle() digunakan untuk menggambar kotak di sekitar wajah. 
    #Parameter pertama adalah gambar asli yang akan dimodifikasi. 
    #Parameter kedua dan ketiga adalah titik koordinat sudut kiri atas kotak yang akan digambar. 
    #Parameter keempat adalah titik koordinat sudut kanan bawah kotak. 
    #Parameter kelima adalah warna kotak dalam format BGR, dengan (0, 255, 0) menghasilkan warna hijau. 
    #Parameter keenam adalah ketebalan garis kotak.
```



*   Kode di atas bertujuan untuk menggambar kotak di sekitar wajah yang telah dideteksi.
*   Setelah langkah ini selesai, kotak akan digambar di sekitar wajah yang telah dideteksi pada gambar.

```python
import cv2
import numpy as np
```

**Menganalisa & menampilkan hasil**

```python
image_path = next(iter(uploaded))
image = cv2.imread(image_path)

face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
faces = face_cascade.detectMultiScale(image, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

for (x, y, w, h) in faces:
    cv2.rectangle(image, (x, y), (x+w, y+h), (0, 255, 0), 2)

from google.colab.patches import cv2_imshow
cv2_imshow(image)
```
